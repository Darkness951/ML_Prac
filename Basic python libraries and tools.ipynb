# Exercise B
Titanic dataset

```python
sns.set_style("whitegrid")
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.barplot(x=df['sex'].unique(), y=df['sex'].value_counts())
```

```python
sns.heatmap(df.corr())
```

```python
print(df.isnull().sum())

# Age has 177 missing values
# Filling them with the median
df["age"].fillna(df["age"].median(), inplace = True)
```

```python
# Plotting class vs. age boxplot
sns.boxplot(x=df["class"], y =df["age"])
```
